# Data Cleaning

```python
df1.isna().sum()
```

```python
cat=[]
con=[]

for i in df1.columns:
    if df1[i].dtypes=="object":
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
for i in df1.columns:
    if df1[i].dtypes=="object":
        m=df1[i].mode()[0]
        df1[i]=df1[i].fillna(m)
    else:
        avg=df1[i].mean()
        df1[i]=df1[i].fillna(avg)
```

```python
df1.isna().sum()
```

# Visual Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))
plt.title("Countplot for Cars / Bar Graph")
sns.countplot(x="Cylinders",data=df1)
plt.grid()


```

```python
sns.histplot(x="Price",data=df1,edgecolor="black",bins=8,color="r")
plt.title("HistPlot for Price OF cars")
plt.xlabel("Price")
plt.ylabel("Cars Count")
plt.grid()
```

```python
p=df1["Type"].value_counts()
plt.title("Pie Chart For Types of Cars")
plt.pie(p,labels=p.index,autopct="%0.02f%%")
```

```python
plt.figure(figsize=(10,6))
plt.title("Scatterplot for Cars")
sns.scatterplot(x="Price",y="Horsepower",data=df1,color='r',hue="Origin",style="Origin")
plt.xlabel("Price")
plt.ylabel("Horsepower")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.title("Boxplot For Cars")
plt.grid()
sns.boxplot(x="Type",y="Price",data=df1,hue="Origin",palette="plasma")
plt.show()
```

```python
#heatmap
#crosstab for to create dataframe
ct=pd.crosstab(df1['Type'],df1['Origin'])
ct
```

```python
plt.title("Heatmap")
sns.heatmap(ct,annot=True)
plt.show()
```

```python
#multivariate

sns.pairplot(df1,x_vars=["EngineSize","Horsepower","MPG.city"],y_vars=["Price"])
plt.suptitle("Mutivariate",y=1.10)
plt.show()
```

```python
plt.grid()
plt.title("Lineplot")
sns.lineplot(x='Min.Price',y='Max.Price',data=df1,hue="Origin",style="Origin")
plt.show()
```
